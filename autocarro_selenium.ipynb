```python
import requests
import selenium
from bs4 import BeautifulSoup
from selenium import webdriver
import time
import pandas as pd
from selenium.webdriver.common.keys import Keys
driver = webdriver.Chrome('/home/sol/notebooks/web_scraper_python/chromedriver')
driver.get('https://autocarro.com.br/auto-busca/carros?AutoBusca=1&qc=&qt=&par=1&q=&ai=&af=&pi=100&pf=500&com=&cam=&cor=&por=&est=&cid=')#start url
```

```python
#lets see the pathern of link adds
links = driver.find_elements_by_tag_name('a')
link_list = []
for link in links:
    link_list.append(link.get_attribute('href'))
```

```python
link_list
```

```python
#filter by a substring
add_list = [link for link in link_list if 'https://autocarro.com.br/particular/anuncio/' in link]
```

```python
add_list
```

```python
#get the unique values
add_list = list(set(add_list))
```

```python
adds = []
contacts = []
for link in add_list:
    driver.get(link)
    driver.maximize_window()
    time.sleep(3)
    add = driver.find_element_by_css_selector('.details-title').text
    contact = driver.find_element_by_tag_name('p').text
    adds.append(add)
    contacts.append(contact)
```

```python
adds, contacts
```

```python
adds_filter = []
for i in adds:
    i = i.replace('\n', ' ')
    adds_filter.append(i)
```

```python
contacts_filter = []
for i in contacts:
    i = i.replace('\n', ' ')
    contacts_filter.append(i)
```

```python
df = pd.DataFrame(list(zip(adds_filter, contacts_filter)), columns=['anuncio', 'contato'])
```
