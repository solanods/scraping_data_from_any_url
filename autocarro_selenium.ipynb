```python
import requests
import selenium
from bs4 import BeautifulSoup
from selenium import webdriver
import time
import pandas as pd
import re
from selenium.webdriver.common.keys import Keys
driver = webdriver.Chrome('/home/sol/notebooks/web_scraper_python/chromedriver')
driver.get('https://autocarro.com.br/auto-busca/carros?AutoBusca=1&qc=&qt=&par=1&q=&ai=&af=&pi=100&pf=500&com=&cam=&cor=&por=&est=&cid=')#start url
```

```python
#lets see the pathern of link adds
links = driver.find_elements_by_tag_name('a')
link_list = []
for link in links:
    link_list.append(link.get_attribute('href'))
```

```python
link_list
```

```python
#filter by a substring
add_list = [link for link in link_list if 'https://autocarro.com.br/particular/anuncio/' in link]
```

```python
add_list
```

```python
#get the unique values
add_list = list(set(add_list))
```

```python
adds = []
contacts = []
for link in add_list:
    driver.get(link)
    driver.maximize_window()
    time.sleep(3)
    add = driver.find_element_by_css_selector('.details-title').text
    contact = driver.find_element_by_tag_name('p').text
    adds.append(add)
    contacts.append(contact)
```

```python
adds, contacts
```

```python
adds
```

```python
#data cleaning
adds = [add.replace('Anncio: ', '')for add in adds] # primeiro remover caracteres indesejados
adds = [add.replace('N de Visitas: ', '')for add in adds] # primeiro remover caracteres indesejados
```

```python
# separar uma lista de strings com base em mais de um separador
adds = [re.split('\n|/ ', add) for add in adds]
```

```python
adds
```

```python
#desempacotar as listas
anuncios = []
modelos =[]
anos =[]
cores =[]
visitantes =[]
for add in adds:
    anuncio, modelo, ano, cor, visitas = add
    anuncios.append(anuncio)
    modelos.append(modelo)
    anos.append(ano)
    cores.append(cor)
    visitantes.append(visitas)
```

```python
# separar uma lista de strings com base em mais de um separador
contacts = [contact.replace('/', '')for contact in contacts] # primeiro remover caracteres indesejados
contacts = [re.split('\n', contact) for contact in contacts] # depois fazer o fatiamento
```

```python
contacts
```

```python
nomes = []
fones =[]
cidade_uf =[]
for contact in contacts:
    nome, fone, cidade = contact
    nomes.append(nome)
    fones.append(fone)
    cidade_uf.append(cidade)
```

```python
df = pd.DataFrame(list(zip(anuncios, modelos, anos, cores, visitantes, nomes, fones, cidade_uf)), columns=['anuncio', 'modelo', 'ano', 'cor', 'visitas', 'nome', 'contatos', 'cidade_uf'])
```
