This script will perform a scraping from https://www.vivareal.com.br.
The goal is obtaining all the links and the phone numbers from house ads with max price 650.000.


```python
#imports
import requests
import selenium
from bs4 import BeautifulSoup
from selenium import webdriver
import time
import pyautogui as pag
```

```python
#instantiate the browser object
driver = webdriver.Chrome('/home/sol/notebooks/web_scraper_python/chromedriver')
```

```python
#open the page
driver.get('https://www.vivareal.com.br/venda/rio-grande-do-sul/novo-hamburgo/bairros/primavera/casa_residencial/#preco-ate=650000')
```

```python
#get the links
links = driver.find_elements_by_tag_name('a')
lista = []
for link in links:
    lista.append(link.get_attribute('href'))
```

```python
lista
```

```python
#removing the none objects and leaving only the strings
lista_so_strings = []
for i in lista:
    if isinstance(i, str):
        lista_so_strings.append(i)
```

```python
lista_so_strings
```

```python
#one more filter
lista_alvo = []
for i in lista_so_strings:
    if len(i) > 128:
        lista_alvo.append(i)
    
```

```python
#get the unique values
lista_alvo2 = list(set(lista_alvo))
```

```python
#loop will traverse all links getting elements by css selector

for i in lista_alvo2:
    driver.get(i)
    driver.maximize_window()
    time.sleep(3)
    #here we need to scroll the page cause Selenium doesn't find a element wich is not visible
    pag.click(2551, 612)
    driver.find_element_by_css_selector('.lead__container .lead-phone__button').click()
    time.sleep(5)
    print(i)
    print(driver.find_element_by_css_selector('.lead__container .post-lead-success__link').text)
    

    
```
